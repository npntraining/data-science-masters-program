# Regression Plots

```python
import seaborn as sns
%matplotlib inline
```

```python
tips_data = sns.load_dataset("tips")
```

```python
tips_data.head()
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips_data)
```

```python
sns_plot=sns.lmplot(x="total_bill",y="tip",data=tips_data,hue="sex")
sns_plot.savefig("output.png")
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips_data,hue="sex",markers=["o","v"])
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips_data,hue="sex",markers=["o","v"],scatter_kws={"s":100})
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips_data,col="sex")
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips_data,col="sex",row='time')
```
